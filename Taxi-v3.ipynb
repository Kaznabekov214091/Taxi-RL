```python
# Re-run this cell to install and import the necessary libraries and load the required variables
!pip install gymnasium[toy_text] imageio
import numpy as np
import gymnasium as gym
import imageio
from IPython.display import Image
from gymnasium.utils import seeding

# Initialize the Taxi-v3 environment
env = gym.make("Taxi-v3", render_mode='rgb_array')

# Seed the environment for reproducibility
env.np_random, _ = seeding.np_random(42)
env.action_space.seed(42)
np.random.seed(42)

# Maximum number of actions per training episode
max_actions = 100 
```

```python
alpha = 0.1
gamma = 0.99
epsilon = 1.0
epsilon_decay = 0.999
min_epsilon = 0.1
max_actions = 100  # Maximum actions per episode
n_states, n_actions = env.observation_space.n, env.action_space.n
episode_returns = []

# Initialize Q-table
q_table = np.zeros((n_states, n_actions))

# Epsilon-greedy policy
def epsilon_greedy(state):
    if np.random.rand() < epsilon:
        action = env.action_space.sample()  # Explore: Random action
    else:
        action = np.argmax(q_table[state, :])  # Exploit: Best action according to Q-table
    return action

# Q-learning update rule
def update_q_table(state, action, reward, next_state):
    best_next_action = np.argmax(q_table[next_state])  # Best next action in the next state
    q_table[state, action] = (1 - alpha) * q_table[state, action] + alpha * (reward + gamma * q_table[next_state, best_next_action])

# Train the agent
for episode in range(2000):
    state, info = env.reset(seed=42)
    terminated = False
    episode_reward = 0
    for t in range(max_actions):  # Limit the number of actions to max_actions
        action = epsilon_greedy(state)  # Choose action based on epsilon-greedy policy
        next_state, reward, terminated, truncated, info = env.step(action)  # Take action and observe result
        episode_reward += reward
        update_q_table(state, action, reward, next_state)  # Update Q-table
        state = next_state
        if terminated:
            break
    episode_returns.append(episode_reward)  # Store total reward for the episode
    epsilon = max(min_epsilon, epsilon * epsilon_decay)  # Decay epsilon

# Extract the learned policy
policy = {state: np.argmax(q_table[state, :]) for state in range(n_states)}

# Test the agent with the learned policy
env.reset(seed=42)
state, info = env.reset(seed=42)
terminated = False
episode_total_reward = 0
frames = []

for t in range(16):  # Limit the number of actions to 16
    action = policy[state]  # Choose action based on the learned policy
    next_state, reward, terminated, truncated, info = env.step(action)  # Take action and observe result
    frames.append(env.render())  # Save the environment state for rendering
    episode_total_reward += reward
    state = next_state
    if terminated:
        break

# Output
print("Total reward for the test episode: ", episode_total_reward)
print("Learned Q-table: \n", q_table)
print("Learned policy: ", policy)
```
